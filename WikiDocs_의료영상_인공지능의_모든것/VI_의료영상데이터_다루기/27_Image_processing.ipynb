```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("/Users/chae/Desktop/TaiLab_Severance_Hospital/dataset/Lung Segmentation Data/Lung Segmentation Data/Train/COVID-19/images/covid_22.png", 0)

 # 히스토그램 평탄화
dst = cv2.equalizeHist(img)
```

```python
 # plot 그리기
plt.figure(figsize=(20,10))

# 히스토그램 평탄화 전
plt.subplot(1,2,1).imshow(img, cmap='gray')
```

```python
# 히스토그램 평탄화 후
plt.subplot(1,2,2).imshow(dst, cmap='gray')
```

```python
# CLAHE 객체 생성 
clahe = cv2.createCLAHE(clipLimit = 3.0, tileGridSize=(12,12))

# CLAHE 적용
img2 = clahe.apply(img)
```

```python
plt.figure(figsize=(20,10))
# 히스토그램 평탄화 전후
plt.subplot(1,3,1).imshow(img, cmap='gray')
plt.subplot(1,3,2).imshow(dst, cmap='gray')


# CLAHE 적용 후 plot 출력 결과물
plt.subplot(1,3,3).imshow(img2, cmap='gray')
```
