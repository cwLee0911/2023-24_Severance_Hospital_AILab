```python
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk
from PIL import Image
```

```python
image_path =  "/Users/chae/Desktop/TaiLab_Severance_Hospital/dataset/Lung Segmentation Data/Lung Segmentation Data/Train/COVID-19/images/covid_83.png"
sitk_image = sitk.ReadImage(image_path)
image_np = sitk.GetArrayFromImage(sitk_image)
```

```python
# 픽셀값의 최소값과 최대값을 구함
min_value = np.min(image_np)
max_value = np.max(image_np)

# Min-Max normalization을 적용
normalized_image = (image_np - min_value) / (max_value - min_value)
```

```python
# 원본과 normalized 된 영상을 plot하여 비교 
plt.figure(figsize=(10,  4))

plt.subplot(1,  2,  1)
plt.imshow(image_np,  cmap='gray')
plt.title('Original Image')

plt.subplot(1,  2,  2)
plt.imshow(normalized_image,  cmap='gray')
plt.title('Normalized Image')
plt.tight_layout()
plt.show()
```

Day 28-3. Z-score Normalization

```python
# 평균값과 표준편차를 계산 
mean_value = np.mean(image_np) 
std_dev = np.std(image_np) 

# Z-score normalization 
normalized_sitk_image = sitk_image - mean_value 
normalized_sitk_image /= std_dev 

# 정규화된 이미지를 다시 numpy array로 변환
normalized_image_np = sitk.GetArrayFromImage(normalized_sitk_image) 
```

```python
# 원본과 normalized 된 영상을 plot하여 비교 
plt.figure(figsize=(10,  4))

plt.subplot(1,  2,  1)
plt.imshow(normalized_image_np,  cmap='gray')
plt.title('Original Image')

plt.subplot(1,  2,  2)
plt.imshow(normalized_image_np,  cmap='gray')
plt.title('Normalized Image')
plt.tight_layout()
plt.show()
```
