Day 13-1. 데이터 정규화 (Normalization)

```python
import pandas as pd
import numpy as np
```

```python
# file in the colab
file_path2 = '/Users/chae/Documents/GitHub/2023-24_Severance_Hospital/WikiDocs_의료영상_인공지능의_모든것/III_정형데이터_다루기/dataset/Heart Failure Clinical Records.csv'

# get tables
df_hf = pd.read_csv(file_path2)
df_hf
```

```python
# MaxAbsScaler 클래스 불러오기
from sklearn.preprocessing import MaxAbsScaler

# df_hf 데이터프레임 'age'와 'platelets' 상위 5개 데이터 출력
dt1= df_hf[[ 'age', 'platelets' ]].head(5)
dt1
```

```python
# MaxAbsScaler 객체 생성 후 abs에 할당
abs = MaxAbsScaler()
# fit 함수로 최대 절대값 계산
abs.fit(dt1)

# transform 함수로 dt1 데이터 스케일링
scaled_dt= abs.transform(dt1)

# 데이터프레임으로 출력
dt_scaled = pd.DataFrame(scaled_dt, columns=dt1.columns)
dt_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler
# MinMaxScaler 객체 생성
minmax = MinMaxScaler()

# fit_transform 함수를 사용하여 dt1 데이터 스케일링
dt_norm = pd.DataFrame(minmax.fit_transform(dt1), columns=dt1.columns)

# 스케일링된 데이터프레임 출력
dt_norm
```

```python
dt1= df_hf[[ 'age' , 'platelets' ]].head(5)
dt1
```

```python
# StandardScaler 불러오기
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성 후 std_s에 할당
std_s = StandardScaler()
std_s

# 평균: 0 , 표준 편차: 1
df_std = pd.DataFrame(std_s.fit_transform(dt1), columns=dt1.columns)
df_std
```

```python
from sklearn.preprocessing import RobustScaler

# RobustScaler로 객체 생성
scaler = RobustScaler()

# fit_transform 함수로 dt1의 데이터 중앙값과 IQR 계산 후 데이터 스케일링
df_robust = pd.DataFrame(scaler.fit_transform(dt1), columns=dt1.columns)
df_robust
```
