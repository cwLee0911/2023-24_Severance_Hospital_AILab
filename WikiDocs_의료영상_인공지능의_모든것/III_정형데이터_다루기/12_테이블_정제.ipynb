```python
import pandas as pd
import numpy as np
```

```python
# Heart csv 파일 데이터프레임 확인
heart_csv = '/Users/chae/Documents/GitHub/2023-24_Severance_Hospital/WikiDocs_의료영상_인공지능의_모든것/III_정형데이터_다루기/dataset/Heart Failure Clinical Records.csv'
df_hf = pd.read_csv( heart_csv )
df_hf
```

```python
df = pd.DataFrame({
    'A': [1, 2, None, 4],
    'B': [5, None, 7, 8]
})

print(df.isnull())
```

```python
# % of missing 을 df_hf 의 모든 column들에 대하여 구함
for col in df_hf.columns:
    pct_missing = np.mean(df_hf[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# 'serum_sodium_median' 새로운 열 생성 후 초기값 NaN 으로 설정
df_hf[ 'serum_sodium_median '] = np.nan
```

```python
df_hf.columns
```

```python
# serum_sodium 와 serum_sodium_median 열의 상위 5값 확인
df_hf[['serum_sodium', 'serum_sodium_median ']].head()
```

```python
# serum_sodium 열의 중앙값 계산 
med = df_hf['serum_sodium'].median()
print (med)
# 출력: 137.0

# serum_sodium_median의 NaN값들을 중앙값으로 대체 (=137.0)
df_hf['serum_sodium_median '] = df_hf['serum_sodium_median '].fillna(med)

# 대체 된 결과값 확인
df_hf[['serum_sodium', 'serum_sodium_median ']].head()
```

```python
df_hf['serum_sodium_median '][1]
```

```python
df_hf[ 'serum_sodium_median '][1] = np.nan
df_hf[ 'serum_sodium_median '].head(3)


# serum_sodium_median 열에서 NaN 값이 있는 행 삭제
df_hf.dropna(subset = ['serum_sodium_median ']).head(3)
```

```python
# 'smoking' 열에서 중복된 행 확인: True or False 반환
df_hf.duplicated(subset=['smoking'])
```
